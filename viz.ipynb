```python
import trimesh
import numpy as np
from keras.models import load_model
```

```python
model = load_model('models/saliency_model-ad-hoc.h5')
model_2 = load_model('models/saliency_model-ad-hoc-2.h5')
```

```python
model_name = 'spider_decimated'
```

```python
mesh = trimesh.load_mesh(f'data-1/{model_name}.obj')
saliency = np.loadtxt(f'data-1/{model_name}.txt')
patches = np.load(f'cached/{model_name}.npy')
```

```python
predications = model.predict(patches)
predications_2 = model_2.predict(patches)
```

```python
mesh.visual.face_colors = trimesh.visual.interpolate(predications, color_map='jet')
mesh.show()
```

```python
mesh.visual.face_colors = trimesh.visual.interpolate(predications_2, color_map='jet')
mesh.show()
```

```python
mesh.visual.face_colors = trimesh.visual.interpolate(saliency, color_map='jet')
mesh.show()
```
