```python
import trimesh
import numpy as np
from keras.models import load_model
```

```python
normalized_model = load_model('models/saliency_model-f32-normalized.h5')
not_normalized_model = load_model('models/saliency_model-f32-not-normalized.h5')
```

```python
model_name = 'shell_decimated'
```

```python
mesh = trimesh.load_mesh(f'data-1/{model_name}.obj')
saliency = np.loadtxt(f'data-1/{model_name}.txt')
patches = np.load(f'cached/{model_name}.npy')
```

```python
patches_normalized = (patches - np.mean(patches, axis=0)) / np.std(patches, axis=0)
```

```python
predications = not_normalized_model.predict(patches)
predications_normalized = normalized_model.predict(patches_normalized)
```

```python
mesh.visual.face_colors = trimesh.visual.interpolate(predications, color_map='jet')
mesh.show()
```

```python
mesh.visual.face_colors = trimesh.visual.interpolate(predications_normalized, color_map='jet')
mesh.show()
```

```python
mesh.visual.face_colors = trimesh.visual.interpolate(saliency, color_map='jet')
mesh.show()
```
